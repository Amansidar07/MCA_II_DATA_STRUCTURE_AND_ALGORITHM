```python
lst = [10,20,30,40,50]
print("original list:", lst)
pos = int(input("enter position to insert(0-based index):"))
element = int (input("enter element to insert:"))
if pos >= 0 and pos <=len(lst):
    lst.insert(pos, element)
    print("list after insertion:", lst)
else:
    print("invalid position!")
pos = int(input("enter position to delete(0-based index):"))
if pos >= 0 and pos < len(lst):
    deleted_element = lst.pop(pos)
    print("deleted element:", deleted_element)
    print("list after deletion:", lst)
else:
    print("invalid position!")
```

```python
lst = [5,15,25,35,45]
print("list:",lst)
key= int(input("enter element to search:"))
if key in lst:
    print("element found at index:",lst.index(key))
else:
    print("element not found!")
```

```python
lst = []
lst.append(10)
lst.append(20)
lst.append(30)

print("after insertion:", lst)
print("traversing list:")
for item in lst:
    print(item, end=" ")
key = 20
if key in lst:
    print("\nelement", key, "found")
else:
    print("\nelement not found")
lst.remove(20)

print("after deletion:", lst)
print("length of list:", len(lst))
```

```python
arr = []
n = int(input("enter number of elements:"))
for i in range(n):
    element= int(input("enter element"))
    arr.append(element)
print("array elements are:")
for i in range(len(arr)):
    print("index",i,"=",arr[i])
```

```python
rows = int(input("enter number of rows:"))
cols = int(input("enter number of columns:"))
matrix = []
for i in range(rows):
    row=[]
    for j in range(cols):
        element = int (input(f"enter element at position ({i},{j}):"))
        row.append(element)
    matrix.append(row)
print("2D array (matrix):")
for i in range(rows):
    for j in range(cols):
        print(matrix[i][j],end=" ")
    print()
```
